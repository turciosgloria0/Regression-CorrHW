```python
#cut = {'Ideal': 1,'Premium': 2,'Very Good':3,'Good':4,'Fair':5}
def score_to_numeric(x):
    if x=='Ideal':
        return 1
    if x=='Premium':
        return 2
    if x=='Very Good':
        return 3
    if x=='Good':
        return 4
    if x=='Fair':
        return 5
    
diamonds_df['CutValue'] = diamonds_df['cut'].apply(score_to_numeric)
diamonds_df.tail(10)
```

```python
#now with cut values I want to see the realtions on cut to price. I will create visuals to check correlation.
corr = diamonds_df.corr()
sns.heatmap(corr, vmin=-1, annot=True)
```

```python
sns.lmplot(x='CutValue', y='depth', data=diamonds_df)
```

```python
#Diamond depth 
#Well cut diamonds = brilliant;
#Deep cut diamonds = Less brilliant; Shallow cut = Less brilliant;
#looking at the graph you can see the diamond "Ideal" tend to be between 58 and 68 depth, while diamonds in the "Fair" have a variance from about 50 to 74 which can lead to a less brilliant diamond.
```

```python
#Diamond colors:['E', 'I', 'J', 'H', 'F', 'G', 'D'] Diamond rankings best to worst based on web 1-7
               #[ 2    6    7    5    3    4    1 ]  
```

```python
def score_to_numeric(x):
    if x=='E':
        return 2
    if x=='I':
        return 6
    if x=='J':
        return 7
    if x=='H':
        return 5
    if x=='F':
        return 3
    if x=='G':
        return 4
    if x=='D':
        return 1
    
diamonds_df['Rankings'] = diamonds_df['color'].apply(score_to_numeric)
diamonds_df.tail(10)
```

```python
diamonds_df.describe() 
```

```python
#Bin prices to four different price groups price range from $326-$18,823
diamonds_df['price'].dtypes
bins=[0,1000,5000,10000,20000]
group_names = ['1', '2', '3', '4']
diamonds_df['PriceLevel'] = pd.cut(diamonds_df['price'], bins, labels=group_names)
diamonds_df.tail()
```

```python
#violin plot to see the diamonds rankings against price levels
sns.violinplot(x='PriceLevel', y='Rankings', data=diamonds_df)
```
